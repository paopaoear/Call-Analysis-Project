In this file, we analyze what the top 10 causes of calls to 311 are in our chosen ZIP code, and calculate how many total incidents of each of these 10 types there have been in the year 2020.

```python
import os
import pandas as pd
```

```python
# import file
cwd = os.getcwd()
file_dir = cwd + '/311_Service_Requests_2020.csv'
df = pd.read_csv(file_dir)
df.head(3)
```

```python
# extract data in our ZIP code (10025)
df1 = df[df['Incident Zip'] == 10025.0]
df1.head(3)
```

```python
# analyze the top 10 causes of calls to 311
top10 = df1['Complaint Type'].value_counts().head(10)
top10
```
