In this file, we analyze whether illegal parking incidents are a larger fraction of total 311 incidents in your ZIP code than they are in general. Specifically, compute the total number of parking incidents in your ZIP, and the total number of all incidents -- is this fraction greater than or smaller than the total fraction of parking incidents across all ZIP codes?

```python
import os
import pandas as pd
```

```python
# import file
cwd = os.getcwd()
file_dir = cwd + '/311_Service_Requests_2020.csv'
df = pd.read_csv(file_dir)
df.head(3)
```

```python
# extract data in our ZIP code (10025)
zip_code = 10025.0
df1 = df[df['Incident Zip'] == zip_code]
df1.head(3)
```

```python
# analyze whether illegal parking incidents are a larger fraction of total 311 incidents in your ZIP code than they are in general
frac0 = df[df['Complaint Type'] == 'Illegal Parking'].shape[0] / df.shape[0]
frac1 = df1[df1['Complaint Type'] == 'Illegal Parking'].shape[0] / df1.shape[0]
higher_parking_proportion = frac1 > frac0
zip_ = int(zip_code)

print(f'higher_parking_proportion: {higher_parking_proportion}\n')
print(f'The fraction of illegal parking in our zip {zip_} is {frac1}, the fraction of that in general is {frac0}.\n\
Thus, it is {higher_parking_proportion} that our ZIP contains a higher proportion of parking incidents than the global value.')
```
